```python
from nltk import sent_tokenize, word_tokenize
from nltk.book import *
import numpy as np
import matplotlib.pyplot as plt


# loading text

x_axis = np.linspace(1, 10, 10)

raw = open('pp.txt',  encoding='utf-8').read()
tokens = word_tokenize(raw)
text10 = Text(tokens)

# metrics

lengths = np.zeros(10, dtype = int)
lexdiv = np.zeros(10, dtype = float)

for i in range(0,10):
    textname = eval('text'+str(i+1))
    lengths[i] = len(textname)
    lexdiv[i] = len(set(textname))/len(textname)

# plots
fig = plt.figure(figsize=(20/2.54,10/2.54), dpi = 80)
ax1 = fig.add_subplot(121)
ax1.plot(x_axis,lengths,'x')
ax1.set_title('size of texts')
plt.grid('both')

ax2 = fig.add_subplot(122)
ax2.plot(x_axis,lexdiv,'x')
ax2.set_title('lexical diversity')
ax2.set_ylim(0,0.150)
plt.grid('both')
plt.show()


# task1
# similar()  и common_contexts - первое вернет набор слов, которые встречаются в похожих контекстах 
# (то есть, обычно вокруг них те же слова, что и вокруг заданных), а второе - общие контексты для переданных слов

# text10.common_contexts(['Elizabeth', 'Mr. Darcy'], num=10)
# text10.concordance('Elizabeth', width=79, lines=5)
text10.dispersion_plot(['Elizabeth', 'Darcy', 'Wickham'])

text10.common_contexts(['Elizabeth', 'Darcy'], num=20)
```
