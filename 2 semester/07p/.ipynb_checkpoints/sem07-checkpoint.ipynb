```python
import numpy as np
import matplotlib.pyplot as plt


x_axis = np.linspace(0, 5, 1000)
sigma = [0.5,1,2]

fig = plt.figure(figsize=(20/2.54,10/2.54), dpi = 80)
ax = fig.add_subplot(111)

for s in sigma:
    y = (1/s)*np.exp(-x_axis/s)

    # plots
    
    
    ax.plot(x_axis,y,label = f'sigma = {s}')
    ax.set_title('size of texts')
    plt.grid('both')

# ax2 = fig.add_subplot(122)
# ax2.plot(x_axis,lexdiv,'x')
# ax2.set_title('lexical diversity')
# ax2.set_ylim(0,0.150)
# plt.grid('both')
ax.legend()
plt.show()
```
